```python
import random

NB_PARTICLES = 10
NB_OPERANDS = 3
MIN_RANGE = 149
MAX_RANGE = 190
TARGET_VALUE = 100
MAX_ITERATIONS = 2000
MAX_VELOCITY = 10
POSITIVE_INFINITY = 100000

position = [];
currentValue = [];
bestValue = []
velocity = [0.0] * NB_PARTICLES
globalBestValue = POSITIVE_INFINITY

def solutionValue(code):
    sum = 0
    for i in range(len(code)):
        sum += code[i]
    return sum

def init(globalBestValue):
    for i in range(NB_PARTICLES):
        code = []
        for j in range(NB_OPERANDS):
            code.append(random.randrange(MIN_RANGE, MAX_RANGE))
        position.append(code)
        currentValue.append(solutionValue(code))
        bestValue.append(currentValue[i])
        if abs(currentValue[i] - TARGET_VALUE) < abs(globalBestValue - TARGET_VALUE):
            globalBestValue = currentValue[i]
    return globalBestValue

def updateVelocity(globalBestValue):
    for i in range(NB_PARTICLES):
        velocity[i] += 2.0 * random.random() * (bestValue[i] - currentValue[i]) \
                    + 2.0 * random.random() * (globalBestValue - currentValue[i])
        if velocity[i] > MAX_VELOCITY:
            velocity[i] = MAX_VELOCITY
        if velocity[i] < -MAX_VELOCITY:
            velocity[i] = -MAX_VELOCITY 
        
def updatePosition(globalBestValue):
    for i in range(NB_PARTICLES):
        for j in range(NB_OPERANDS):
            position[i][j] += int(velocity[i])
        currentValue[i] = solutionValue(position[i])
        if abs(currentValue[i] - TARGET_VALUE) < abs(bestValue[i] - TARGET_VALUE):
            bestValue[i] = currentValue[i]
        if abs(currentValue[i] - TARGET_VALUE) < abs(globalBestValue - TARGET_VALUE):
            globalBestValue = currentValue[i]
    return globalBestValue

def printCurrent(iteration):
    print('Iteration', iteration)
    for i in range(NB_PARTICLES):
        for j in range(NB_OPERANDS):
            if j < NB_OPERANDS - 1:
                print(position[i][j], '+', end=' ')
            else:
                print(position[i][j], '=', end=' ')
        print(currentValue[i])
    print()
    
def PSO(globalBestValue):
    globalBestValue = init(globalBestValue)
    for iteration in range(MAX_ITERATIONS):
        if iteration % 10 == 0:
            printCurrent(iteration)
        if globalBestValue == TARGET_VALUE:
            print('Solution found after', iteration, 'iterations.')
            return
        updateVelocity(globalBestValue)
        globalBestValue = updatePosition(globalBestValue)
    print('Solution not found')
    
if __name__ == '__main__':
    PSO(globalBestValue)
    
```
